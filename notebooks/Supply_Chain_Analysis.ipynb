## 1. The Business Problem
In retail, maintaining a high **Service Level** (98%+) while minimizing **Holding Costs** is a classic optimization challenge. This notebook walkthrough demonstrates how to:
1. Forecast SKU-level demand.
2. Calculate mathematical **Safety Stock** and **Reorder Points**.
3. Identify **Stockout Risks** before they happen.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the pre-generated data
sales = pd.read_csv('../data/sales_history.csv')
meta = pd.read_csv('../data/sku_metadata.csv')
results = pd.read_csv('../data/inventory_recommendations.csv')

print(f"Loaded {len(sales)} sales records across {len(meta)} SKUs.")
```

## 2. Demand Seasonality Analysis
We analyze patterns across days of the week and months to understand when the supply chain is most stressed.

```python
sales['date'] = pd.to_datetime(sales['date'])
sales['weekday'] = sales['date'].dt.day_name()
sales_pivot = sales.pivot_table(index='weekday', values='sales', aggfunc='mean')
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_pivot = sales_pivot.reindex(days)

plt.figure(figsize=(10, 4))
sns.lineplot(data=sales_pivot, marker='o', color='teal')
plt.title('Average Daily Sales Volume by Day of Week')
plt.grid(True, alpha=0.3)
plt.show()
```

## 3. Inventory Optimization Math
We use the **Normal Distribution** to calculate Safety Stock:

$$Safety Stock = Z \times \sigma_{demand} \times \sqrt{LeadTime}$$

Where:
*   **Z** = Service level factor (1.645 for 95%)
*   **$\sigma$** = Standard deviation of demand
*   **LeadTime** = Time from order to delivery

```python
# Show the top 10 recommended reorder points
print("Top 10 SKUs by Reorder Point Requirement:")
display(results[['sku_id', 'avg_daily_demand', 'avg_lead_time_days', 'safety_stock', 'reorder_point', 'eoq']].head(10))
```
